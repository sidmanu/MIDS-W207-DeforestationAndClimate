```python
carbon_average = carbon.groupby(by='year').mean()
```

```python
carbon_average = carbon_average.drop('Unnamed: 0', axis=1)
carbon_average = carbon_average.rename(columns={'gfw_forest_carbon_gross_emissions__Mg_CO2e': 'average'})
```

```python
carbon.columns
```

```python
carbon = carbon.pivot(index='year', columns='country', values='gfw_forest_carbon_gross_emissions__Mg_CO2e')
carbon = pd.merge(carbon, carbon_average, on = 'year')
carbon = carbon.rename_axis(columns=None).reset_index()
carbon.columns = ["_".join(('carbon',i)) for i in carbon.columns]
carbon.head(5)
```

```python
tc_average = tc.groupby(by='year').mean()
```

```python
tc_average = tc_average.drop(['Unnamed: 0', 'threshold','gain_2000-2020_ha','area_ha','extent_2000_ha','extent_2010_ha'], axis=1)
tc_average = tc_average.rename(columns={'tc_loss_ha':'tc_loss_ha_average'})
```

```python
tc = tc.pivot(index='year', columns='country', values=['area_ha','extent_2000_ha','extent_2010_ha','tc_loss_ha'])
tc.columns = ["_".join((i,j)) for i,j in tc.columns]
tc = pd.merge(tc,tc_average, on ='year')
tc.reset_index()

tc.head(5)
```

```python
carbon=carbon.loc[:, (carbon != 0).any(axis=0)]
tc=tc.loc[:, (tc != 0).any(axis=0)]
```

```python
df = pd.merge(df, carbon, left_on = 'year', right_on = 'carbon_year')
df = pd.merge(df, tc, left_on = 'year', right_on = 'year')
```

```python
train_df = df[(df.year <= 2014)]
validation_df = df[((df.year <= 2018) & (df.year >= 2015))]
test_df = df[(df.year >= 2019)]
```
