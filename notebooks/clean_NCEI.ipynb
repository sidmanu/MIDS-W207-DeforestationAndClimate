```python
import pandas as pd
import numpy as np

df = pd.read_csv('USC00048758.csv')
```

```python
df.columns = map(str.lower, df.columns)

print(df.shape)
print(df.columns)

df.head(5)
```

```python
print(df.dtypes)
```

```python
c = ['STATION','ELEVATION','NAME','LATITUDE','LONGITUDE','CDSD_ATTRIBUTES','DSND_ATTRIBUTES','DYFG_ATTRIBUTES','DYTS_ATTRIBUTES']
c = [x.lower() for x in c]

df = df.drop(columns=c)
```

```python
for column in df.columns:
    if df[column].dtype == int or df[column].dtype == float:
        pass
    else:
        new_column = []
        for val in df[column]:
            if isinstance(val, float) or isinstance(val, int):
                new_column.append(val)
            elif isinstance(val, str) and val.isnumeric():
                new_column.append(float(val))
            elif "," in val:
                sub_list = []
                split_str = val.split(',')
                for s in split_str:
                    if s == '':
                        sub_list.append(None)
                    elif s.isnumeric():
                        sub_list.append(float(s))
                    else:
                        sub_list.append(s)
                new_column.append(sub_list)
            else:
                new_column.append(None)
        df[column] = new_column
```

```python
df.to_csv('clean_NCEI_tahoe.csv', sep='\t')
```
