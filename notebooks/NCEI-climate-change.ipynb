# Download data


NCEI Direct Download link -> https://www.ncei.noaa.gov/metadata/geoportal/rest/metadata/item/gov.noaa.ncdc:C00947/html 


First download the archive file and expand it in the data/NCEI directory.

```python
import pandas as pd
import numpy as np
import os
```

```python
LIMIT = 10

def load_data(path_to_data):
    


    dataframes = []
    count = 0
    
    for fname in os.listdir(path_to_data):

        count += 1
        if count >= LIMIT:
            break

        if not fname.endswith('.csv'):
            continue 

        country_name = fname[:2]
        print("FIPS country code:", country_name)

        data_file_path = os.path.join(path_to_data, fname)

        df = pd.read_csv(data_file_path)
        df['country_name_FIPS'] = country_name
        dataframes.append(df)
    


    return dataframes

```

```python

DATA_PATH = 'data/NCEI/gsoy-latest'
dfs = load_data(DATA_PATH)
```
